```python
import pandas as pd
```

```python
gene_names = pd.read_csv("/oak/stanford/groups/horence/JuliaO/nf-readzs/visium/V1_Mouse_Kidney/annotated_files/annotated_windows.file",sep="\t")
```

```python
gene_name_dict = {k : v for k, v in zip(gene_names["window"], gene_names["gene"])}
```

```python
score1 = "ReadZS_norm"
samples = pd.read_csv("/oak/stanford/groups/horence/JuliaO/visium_analysis/notebooks/output/make_samplesheet/spatial.csv",index_col = 0)
outpath = "/oak/stanford/groups/horence/JuliaO/visium_analysis/notebooks/output/GO_files/"
pval_col = "perm_pvals_emp_adj"
# plot_col = "perm_pvals_emp_adj"
plot_col = "score_cont"
# plot_col = pval_col
scores = pd.read_csv("output/make_samplesheet/scores.csv",index_col=0)
srow = scores.loc[score1]
datanames = ['V1_Mouse_Brain_Sagittal_Posterior','V1_Mouse_Brain_Sagittal_Posterior_Section_2',
       'V1_Mouse_Brain_Sagittal_Anterior',
       'V1_Mouse_Brain_Sagittal_Anterior_Section_2', 'V1_Mouse_Kidney']
```

```python
for dataname in datanames[:1]:
  print(dataname)
  row1 = samples.loc[ dataname]
  df1 = pd.read_csv(row1[srow["ising"]],sep="\t")
  df1["gene"] = df1["window"].map(gene_name_dict)
  GO_file = open("{}{}_{}_GO_sig.txt".format(outpath, dataname, score1),"w")
  s = ""
  for gene in df1[df1[pval_col] < 0.05]["gene"]:
    
    if len(gene) > 1:
      GO_file.write(gene + "\n")
      s += gene + "\n"
  print("sig")
  print(s)
  GO_file.close()
  GO_file = open("{}{}_{}_GO_all.txt".format(outpath, dataname, score1),"w")
  s = ""
  for gene in df1["gene"]:
    if len(gene) > 1:
      GO_file.write(gene + "\n")
      s += gene + "\n"
  print("ref")
  print(s)
  GO_file.close()
```
