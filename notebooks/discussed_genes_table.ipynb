```python
from os.path import exists
import pandas as pd
```

```python
outpath = "/oak/stanford/groups/horence/JuliaO/visium_analysis/notebooks/output/discussed_genes_table/"
samples = pd.read_csv("/oak/stanford/groups/horence/JuliaO/visium_analysis/notebooks/output/make_samplesheet/spatial.csv",index_col = 0)
scores = pd.read_csv("output/make_samplesheet/scores.csv",index_col=0)
```

```python
pval_col = "perm_pvals_emp_adj"
out = {"dataname" : [], "score" : [], "gene" : [], pval_col : [], "xth_sig" : []}#, "sig_by_ge" : []}

name_dict = {"chr11_10983_plus" : "Gpx3", "chr17_16275_minus" : "Slc8a1", "chr8_11013_plus" : "Gpm6a", "chr5_28581_minus" : "Actb", "chr4_23431_minus" : "Rps8", "chr1_30673_plus" : "S100a9"}

genes = ["Myl6", "Rps24", "Gng13","chr11_10983_plus", "chr17_16275_minus", "chr8_11013_plus", "chr5_28581_minus", "chr4_23431_minus", "chr1_30673_plus"]

for gene in genes:
  if gene not in name_dict.keys():
    name_dict[gene] = gene

eff_size = 0
pval = 0.05



for dataname, row in samples[samples["method"] == "visium"].iterrows():
  for score, srow in scores.iterrows():
    if exists(row[srow["ising"]]):

      df = pd.read_csv(row[srow["ising"]],sep="\t")
#       df = df.sort_values("score_cont",ascending=False)
      df = df.sort_values([pval_col, "score_cont"], ascending = [True, False])
      df = df.reset_index(drop=True)
      display(df.head())
      print(dataname)
      if pval_col in df.columns:
        for gene in genes:
          if gene in df[srow["genecol"]].unique():
            gene_df = df[df[srow["genecol"]] == gene]
            out["dataname"].append(dataname)
            out["score"].append(score)
            out["gene"].append(name_dict[gene])
            out[pval_col].append(gene_df[pval_col].iloc[0])
            out["xth_sig"].append(gene_df.index[0] + 1)
#             out[]
#         out["num_genes"].append(df[srow["genecol"]].nunique())
#         out["num_sig"].append(df[(df[pval_col] < pval) & (df["score_cont"] > eff_size)][srow["genecol"]].nunique())
#         df["inv_pval"] = 1 - df[pval_col]
#         top = list(df[(df[pval_col] < pval) & (df["score_cont"] > eff_size)].sort_values(["inv_pval","score_cont"],ascending=False)[srow["genecol"]].head(10))
#         out["top"].append(", ".join(top))
outdf = pd.DataFrame.from_dict(out)
outdf["sig"] = False
outdf.loc[outdf[pval_col] < pval, "sig"] = True
outdf["gene_dataset"] = outdf["gene"] + "_" + outdf["dataname"]
```

```python
temp = outdf[outdf["score"] == "ReadZS_norm"].merge(outdf[outdf["score"] == "ReadZS_ge_norm"][["sig","gene_dataset"]],on="gene_dataset",suffixes = ["", "_by_ge"])
temp = temp.sort_values(["gene", pval_col])
temp.to_csv("{}/readzs_table.csv".format(outpath),index=False)
temp
```
